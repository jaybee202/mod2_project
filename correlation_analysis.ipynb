This notebook includes initial correlation analysis of the kc_house_data.csv data.  In regards to the missing data, alternate analysis is required to justify any exclusion of data.

### Import required packages

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

### Import data and drop columns not required for correlation analysis
These columns include:
- price (the dependant variable in this analyis)
- id is arbutrary to the price of a house
- zipcode (there are 70 zipcodes, too many categories for dummy data)
- Given that the sqft_living and sqft_above subtly measure the same thing, sqft_above is dropped in favor of the total living space instead

```python
data = pd.read_csv('data_cleaned.csv', index_col=0)
data = data.drop(columns=['price', 'id', 'zipcode', 'long', 'lat', 'sqft_above'])
```

```python
corr_df = data.drop(columns=['waterfront', 'yr_renovated'])
corr_df = data.dropna(axis=0)
```

```python
corr_df = corr_df.corr()
```
